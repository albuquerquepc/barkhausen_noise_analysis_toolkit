```python
import matplotlib.pyplot as plt
import numpy as np

path_freq: str = "/home/paulo/Documentos/ic_gmag/medidas/Py(1000nm)/Barkhausen/Power_Spectrum_0.17s_0.22s_Py_1000nm_R798D_0.05Hz_100kHz_4MSs/Frequencies.dat"
spectrum: str = "/home/paulo/Documentos/ic_gmag/medidas/Py(1000nm)/Barkhausen/Power_Spectrum_0.17s_0.22s_Py_1000nm_R798D_0.05Hz_100kHz_4MSs/Power_Spectrum.dat"

x = np.loadtxt(path_freq)
y = np.loadtxt(spectrum)

plt.loglog(x, y, linestyle="none", marker='o', ms=1, color="black")
plt.xlabel("Frequência (Hz)")
plt.ylabel("S(f) (arb. unit)")
plt.xlim(1, max(x)+10e5)
#plt.savefig(f"/home/paulo/Área de trabalho/", dpi=600)
plt.show()
```
